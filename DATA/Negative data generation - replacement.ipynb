## Replacement

```python
def exchange_segments(dna_sequence, num_segments=25, segment_length=24, num_exchanges=9):
    """
    Exchanges segments of a DNA sequence and shuffles their nucleotides.

    Parameters:
    dna_sequence (str): The original DNA sequence to be modified.
    num_segments (int): The total number of segments to divide the sequence into (default is 25).
    segment_length (int): The length of each segment (default is 24).
    num_exchanges (int): The number of segments to exchange (default is 9).

    Returns:
    str: A new DNA sequence with exchanged segments and shuffled nucleotides.
    """
    # Divide the sequence into segments
    segments = [dna_sequence[i:i+segment_length] for i in range(0, len(dna_sequence), segment_length)]

    # Randomly select segments to exchange
    segments_to_exchange = random.sample(segments, num_exchanges)

    # Shuffle the order of segments
    random.shuffle(segments_to_exchange)
    # Swap the selected segments
    for i in range(num_exchanges):
        segments[i], segments_to_exchange[i] = segments_to_exchange[i], segments[i]

    # Shuffle the nucleotides within each moved segment
    for i in range(num_exchanges):
        segments[i] = ''.join(random.sample(segments[i], len(segments[i])))

    # Concatenate segments back into a single DNA sequence
    new_dna_sequence = ''.join(segments)

    return new_dna_sequence
```

```python
non_tata_neg_df = []

# Iterate through each row in df_non_tata
for idx, row in df_non_tata.iterrows():
    # Get the record from the current row
    record = row['seq']

    # Generate the negative record using your function
    negative_record = exchange_segments(record, num_segments=25, segment_length=24, num_exchanges=8)

    # Append the new data to the list
    non_tata_neg_df.append({'ID': row['ID'],'lengths':600, 'seq': negative_record, 'organism':row['organism']})

# Create a new DataFrame from the collected data
non_tata_neg_df = pd.DataFrame(non_tata_neg_df)
```
